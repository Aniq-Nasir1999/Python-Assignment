```python
class Car:
    def __init__(self, company, fuel):
        self.company = company
        self.name = 'COROLA'
        self.colour = 'WHITE'
        self.transmission = 'AUTOMATIC'
        self.fuel = fuel
        
    def updatename(self, newname):
        self.name = newname

    def updatecolour(self, newcolour):
        self.colour = newcolour
        
    def updatetransmission(self, newtransmission):
        self.transmission = newtransmission
        
car1 = Car('TOYOTA', 'DIESEL')
print(car1.company)
print(car1.name)
print(car1.colour)
print(car1.transmission)
print(car1.fuel)
print('\n')
car2 = Car('HONDA', 'PETROL')
print(car2.company)
car2.updatename('CIVIC')
print(car2.name)
car2.updatecolour('BLACK')
print(car2.colour)
car2.updatetransmission('MANUAL')
print(car2.transmission)
print(car2.fuel)
print('\n')
car3 = Car('SUZUKI', 'PETROL')
print(car3.company)
car3.updatename('SWIFT')
print(car3.name)
car3.updatecolour('RED')
print(car3.colour)
car3.updatetransmission('MANUAL')
print(car3.transmission)
print(car3.fuel)
print('\n')
car4 = Car('NISSAN', 'PETROL')
print(car4.company)
car4.updatename('SUNNY')
print(car4.name)
car4.updatecolour('BLUE')
print(car4.colour)
car4.updatetransmission('MANUAL')
print(car4.transmission)
print(car4.fuel)
print('\n')
car5 = Car('MITSUBISHI', 'PETROL')
print(car5.company)
car5.updatename('LANCER')
print(car5.name)
car5.updatecolour('GREEN')
print(car5.colour)
car5.updatetransmission('AUTOMATIC')
print(car5.transmission)
print(car5.fuel)
```

```python
OBJECT ORIENTED PROGRAMMING
Object-oriented programming (OOP) is a programming language model in
which programs are organized around data, or objects, rather than 
functions and logic. An object can be defined as a data field that
has unique attributes and behavior. Examples of an object can range from 
physical entities, such as a human being that is described by properties
like name and address,down to small computer programs, such as widgets.
```

```python
BENEFITS OF OOP
Some common benefits of OOP are listed below:
    1) Modularity for easier troubleshooting. 
    2) Reuse of code through inheritance. 
    3) Flexibility through polymorphism. 
    4) Effective problem solving.
```

```python
DIFFERENCE BETWEEN FUNCTION AND METHOD
A very general and main difference between a "function" and
a "method" is that functions are defined outside of classes, 
while methods are defined inside of and part of classes.
```

```python
CLASS
It is a template or blueprint about the capability of what an object can do.

OBJECT
An object is an element (or instance) of a class, objects have the behaviors 
of their class.

ATTRIBUTE
Attributes are data stored inside a class or instance and represent the state
or quality of the class or instance. In short, attributes store information about the instance.

BEHAVIOUR
A class's behavior determines how an instance of that class operates, 
for example, how it will react if asked to do something by another class or object or
if its internal state changes. Behavior is the only way objects can do anything to
themselves or have anything done to them.
```
